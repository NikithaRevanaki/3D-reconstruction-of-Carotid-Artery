```python
z_cordinate=[]
x_cordinate=[]
s=1.5
y_axis=[]
for i in range(2,40,2):

  z=[]
  x=[]
  img=np.zeros((l,w,3), np.uint8)
  img[::]=(255,255,255)
  cv2.circle(img, (w//2,l//2), i, (255,0,0), thickness=1, lineType=8, shift=0)
  for k in range(0,l):
    for j in range(0,w):
      b,g,r=img[k,j]
      if (b,g,r)!=(255,255,255):
        z.append(j)
        x.append(k)
  #z.append(h)
  #x.append(w)
  z_cordinate.append(z)
  x_cordinate.append(x)
  y_axis.append(s)
  s=s+1.5
```
